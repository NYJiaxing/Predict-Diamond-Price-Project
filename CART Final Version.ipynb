## CART (Regression Tree) Method

```python
import numpy as np
import pandas as pd
data=pd.read_csv("Diamond Dataset.csv")
data.head()
data.loc[data.Color == 'P','Color'] = 0
data.loc[data.Color == 'M','Color'] = 1
data.loc[data.Color == 'L','Color'] = 2
data.loc[data.Color == 'K','Color'] = 3
data.loc[data.Color == 'J','Color'] = 4
data.loc[data.Color == 'I','Color'] = 5
data.loc[data.Color == 'H','Color'] = 6
data.loc[data.Color == 'G','Color'] = 7
data.loc[data.Color == 'F','Color'] = 8
data.loc[data.Color == 'E','Color'] = 9
data.loc[data.Color == 'D','Color'] = 10
data.loc[data.Clarity == 'I3','Clarity'] = 0
data.loc[data.Clarity == 'I2','Clarity'] = 1
data.loc[data.Clarity == 'I1','Clarity'] = 2
data.loc[data.Clarity == 'SI2','Clarity'] = 3
data.loc[data.Clarity == 'SI1','Clarity'] = 4
data.loc[data.Clarity == 'VS2','Clarity'] = 5
data.loc[data.Clarity == 'VS1','Clarity'] = 6
data.loc[data.Clarity == 'VVS2','Clarity'] = 7
data.loc[data.Clarity == 'VVS1','Clarity'] = 8
data.loc[data.Clarity == 'IF','Clarity'] = 9
data.loc[data.Clarity == 'FL','Clarity'] = 10
data.loc[data.Cut == 'Fair','Cut'] = 0
data.loc[data.Cut == 'Good','Cut'] = 1
data.loc[data.Cut == 'Very Good','Cut'] = 2
data.loc[data.Cut == 'Excellent','Cut'] = 3
data.loc[data.Symmetry == 'Fair','Symmetry'] = 0
data.loc[data.Symmetry == 'Good','Symmetry'] = 1
data.loc[data.Symmetry == 'Very Good','Symmetry'] = 2
data.loc[data.Symmetry == 'Excellent','Symmetry'] = 3
data['Cert'] = ['GIA' if x == 'GIA' else 'Other' for x in data['Cert']]
data.loc[data.Cert == 'Other','Cert'] = 0
data.loc[data.Cert == 'GIA','Cert'] = 1
def dummy_df(df, todummy_list):
    for x in todummy_list:
        dummies = pd.get_dummies(df[x], prefix=x, dummy_na=False)
        df = df.drop(x, 1)
        df = pd.concat([df, dummies], axis=1)
    return df
todummy_list = ['Shape']
data = dummy_df(data, todummy_list)
print(data.head())
```

```python
data.head()
```

```python
data = data.dropna()
```

```python
cols = list(data.columns.values)
cols.pop(cols.index('Price'))
data = data[cols + ['Price']]

X = data.drop(['Price'],1)
#data.Price = np.log(data.Price)
Y = data['Price']

data.Depth = pd.Series(data.Depth).convert_objects(convert_numeric=True)
data.Table = pd.Series(data.Table).convert_objects(convert_numeric=True)

X.head()
```

```python

from sklearn.cross_validation import train_test_split

X_train,X_test, y_train, y_test = train_test_split(X,Y,test_size=0.35, random_state=0) 
```

```python
# convert price data to classes

n_classes = 6

y_classes = np.linspace(0, Y.quantile(0.95), n_classes)
y_train_cl = np.digitize(y_train, bins= y_classes)
y_test_cl = np.digitize(y_test, bins= y_classes)

print('Price classes: %s' % (y_classes))
print(Y.quantile(0.95))
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
clf = DecisionTreeClassifier(criterion= 'gini', random_state = 0,  max_depth = 5)
```

```python
clf.fit(X_train,y_train_cl)
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, KFold
kfold = KFold(n_splits=10, random_state=123)
cv_results = cross_val_score(clf, X_train, y_train_cl, cv=kfold, n_jobs=-1) 
print("Predict Result mean", cv_results.mean())
```

```python
from sklearn.externals.six import StringIO  
from sklearn import tree
from sklearn.tree import _tree
import pydotplus
export_graphviz(clf, feature_names = X_train.columns)
dotfile = StringIO()
tree.export_graphviz(clf, out_file=dotfile, feature_names= X_test.columns) 
#generate decision tree diagram
pydotplus.graph_from_dot_data(dotfile.getvalue()).write_png("my_tree1.png")
```

```python
from sklearn import metrics
print("R2 score is",r2_score(y_test_cl, clf.predict(X_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
cclf = RandomForestClassifier(n_estimators=100, criterion= 'gini', random_state = 0,  max_depth = 13)
cclf.fit(X_train, y_train_cl)

metrics.mean_absolute_error(y_test, cclf.predict(X_test))
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, KFold
kfold = KFold(n_splits=10, random_state=123)
cv_results = cross_val_score(cclf, X_train, y_train_cl, cv=kfold, n_jobs=-1) 
print("Predict result mean",cv_results.mean())
print("Predict result standard deviation",cv_results.std())
```
